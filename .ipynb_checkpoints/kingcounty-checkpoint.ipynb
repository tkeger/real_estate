## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Dataset

```python
df = pd.read_csv('kc_house_data.csv', parse_dates=True)
```

```python
df.head().T
```

## Features

**date** = Sale date

**price** = Selling price

**bedrooms** = No. of bedrooms

**bathrooms** = No. of bathrooms (0.75 = bathroom with either shower or bath instead of both)

**sqft_living** = Square footage of the living space

**sqft_lot** = Square footage of the land lot

**floors** = No. of floors

**waterfront** = Whether house has a waterfront (1) or not (0)

**view** = View of the property rated from 0 to 4

**condition** = Condition of the object rated from 1 to 5

**grade** = Rating of construction from 1 to 13

**sqft_above** = Square footage above ground level

**sqft_basement** = Square footage below ground level

**yr_built** = Year property was built

**yr_renovated** = Last renovation

**zipcode** = Zipcode of the property

**lat** = Geographic latitude

**long** = Longitude

**sqft_living15** = Square footage of living space of the 15 nearest neighbors

**sqft_lot15** = Square footage of land lots of the 15 nearest neighbors

### Feature expansion

*We have to get rid of the outliers identified above*

```python
df_out = df[df.price < df.price.quantile(q=0.95)]
```

```python
#How many outliers have been removed?
len(df) - len(df_out)
```

#### Add seasonal features

```python
#Add month
```

```python
#First make a copy of the dataframe
df_new = df_out.copy()
```

```python
df_new['month'] = pd.DatetimeIndex(df_new['date']).month
```

```python
df_new['month']
```

```python
#Make a visualization of total volumes in each month
df_new.groupby(['month']).sum().plot.bar(y = ['price'])
```

```python
#Add seasons to dataframe (0,1,2,3 representing winter, summer, spring, fall)
conditions = [
    (df_new['month'] <3) | (df_new['month'] == 12),
    (df_new['month'] >2) & (df_new['month'] < 6),
    (df_new['month'] >5) & (df_new['month'] < 9),
    (df_new['month'] >8) & (df_new['month'] < 12)
]

choices = [0,1,2,3]
```

```python
df_new['season']= np.select(conditions, choices)
```

```python
#One-hot encode seasons
```

```python
df_new[['Winter','Spring','Summer','Fall' ]] = pd.get_dummies(df_new['season'])
```

```python
df_new = df_new.drop(['Winter','season','month'],axis=1)
```

### Preprocessing

#### Train/Test split


```python
from sklearn.model_selection import train_test_split
```

```python
#Target variable = Price
```

```python
y = df_new['price']
```

```python
X = df_new.drop(['id','price','date'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train.astype(np.float))
X_test = scaler.transform(X_test.astype(np.float))
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
#Predition
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
df_pred = pd.DataFrame({'Actual':y_test,'Predicted': y_pred})
```

```python
#Change scientific format to float for aesthetic reasons:
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df_pred
```

## Evaluation

```python
from sklearn import metrics
```

```python
#Mean Absolute Error:
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
#Mean Squared Error:
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
#Root Mean Squared Error:
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print("test score  :", lr.score(X_test, y_test))
```
